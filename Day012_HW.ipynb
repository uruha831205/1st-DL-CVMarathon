### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find_all('span')[-1].text,d.find_all('em')[-1].text, d.find_all('a')[-1].text)
    
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
dic = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    #print(d.find_all('span')[-1].text,d.find_all('em')[-1].text, d.find_all('a')[-1].text)
    #dic.update({d.find_all('em')[-1].text:d.find_all('a')[-1].text.replace('\u3000','')})
    #tag = d.select('.tag')[0].text
    tag = d.find_all('em')[0].text
    #print(tag,"*")
    dic.setdefault(tag,[])
    dic[tag].append({'title':d.find_all('a')[-1].text.replace('\u3000',''),'date':d.find_all('span')[-1].text})
    #print(dic)
print(dic)
```
